```python
#Exemplo Pratico While
contador = 1
while contador <= 3:
    print("Contagem: ", contador)
    contador += 1
```

```python
#Exemplo prático do for:
frutas = ["maça", "Banana", "Uva"]
for fruta in frutas:
    print("Eu gosto de:", fruta)
```

```python
compras = ["pão", "leite", "café"] # Os índices começam no zero! O pão é o índice 0.
print(compras)     # Vai exibir "pão"
compras.append("ovo") # O '.append' adiciona "ovo" no final da lista
print(compras)        # Exibe: ['pão', 'leite', 'café', 'ovo']
```

```python
# 'Ensinando' a função (definindo os parâmetros):
def calcular_dobro(numero):
    resultado = numero * 2
    return resultado # É o valor que a função "devolve" para você

# Usando a função na prática:
meu_valor = calcular_dobro(5)
print(meu_valor) # Vai exibir 10
```
